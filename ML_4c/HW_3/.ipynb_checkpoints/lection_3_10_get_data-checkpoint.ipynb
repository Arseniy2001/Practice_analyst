### read_parquet (read_json, read_excel, read_sql)

```python
parquetName = 'bigDataExample.parquet'
csvName = 'bigDataExample.csv'
bigData = pd.read_parquet(parquetName)
bigData.head(5)
```

#### Сохранение в новый формат

```python
bigData.to_csv(csvName)
```

```python
from pathlib import Path
pSize = Path(parquetName).stat().st_size
cSize = Path(csvName).stat().st_size
print('parqSize=',pSize, 'csvSize=',cSize, cSize/pSize)
```
