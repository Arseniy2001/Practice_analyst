```python
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as p
%matplotlib inline
```

#### Загружаем данные, разбиваем на обучающую и проверяющие выборки

```python
wine = datasets.load_wine()
XX = p.DataFrame(wine.data[:, [7, 12]], columns=['Нефлаваноидные фенолы','Пролин'])
yy = wine.target
X_train, X_test, y_train, y_test = train_test_split(XX, yy, test_size=0.3)
X_train.hist()
```

#### С помощью метода fit_transform "обучаем" StandardScaler и сразу же делаем преобразование

```python
scaler = StandardScaler()
scaled = p.DataFrame(scaler.fit_transform(X_train), columns=['Нефлаваноидные фенолы','Пролин'])
scaled.hist()
```

#### данные о мат. ожиданиях и дисперсиях по каждому из признаков

```python
np.set_printoptions(suppress=True)#не отображать с e+00
print("mean={} std={}".format(scaler.mean_,scaler.scale_))
```

#### Новые данные. Вызываем только transform, так как данные о среднем значении и отклонении уже есть. Приводим к ним

```python
scaled_test = p.DataFrame(scaler.transform(X_test), columns=['Нефлаваноидные фенолы','Пролин'])
scaled_test.hist()
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
X_train.hist()
minmaxscaler = MinMaxScaler()#сохраняет информацию об изначальном распределении данных
minmaxscaled = p.DataFrame(minmaxscaler.fit_transform(X_train), columns=['Нефлаваноидные фенолы','Пролин'])
minmaxscaled.hist()
```

## Normalizer

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
normalized = p.DataFrame(normalizer.fit_transform(X_train), columns=['Нефлаваноидные фенолы','Пролин'])
normalized.hist()
```
