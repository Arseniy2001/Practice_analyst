```python
import numpy as np
import numpy.random as r
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Сгенерируем данные как $y = 3X^2 + 2X + 5 + $белый шум

```python
X = 3 * r.rand(300,1) - 1
```

```python
y = 3*X*X + 2*X + 5 + r.randn(300,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.333, random_state=1)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model.fit(X_train, y_train)
y_predPipe = model.predict(X_test)
```

Residuals = expected - predicted

```python
residuals = y_test - y_predPipe
```

### 1. Гомоскедастичность – постоянство дисперсии

```python
plt.plot(residuals, "b.")
plt.hlines(-1.7, 0, 100)
plt.hlines(1.7, 0, 100)
```

### 2. Проверка нормальной распределенности

```python
from pandas import DataFrame
df = DataFrame(residuals)
```

```python
df.hist(bins=10)
```

```python
df.plot(kind='kde', legend = False) 
#kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable
```

Более надежный вариант показать, что распределение нормальное - построить график QQ (квантиль-квантиль). Это график, на котором квантили из двух распределений расположены относительно друг друга, тем самым мы сравнивем наше распределение с другим распределением, в данном случае с нормальным. Если два сравниваемых распределения идентичны, график Q – Q следует линии y = x под углом 45°

https://en.wikipedia.org/wiki/Q-Q_plot
https://desktop.arcgis.com/ru/arcmap/latest/extensions/geostatistical-analyst/normal-qq-plot-and-general-qq-plot.htm

```python
import scipy
residulas1dim = residuals[:,0]
qq = scipy.stats.probplot(residulas1dim, dist="norm", plot=plt)
```

Для сравнения тот же график QQ для нормального распределения

```python
qq2norm = scipy.stats.probplot(r.randn(100,1)[:,0], dist="norm", plot=plt)
```
