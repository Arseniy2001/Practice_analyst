```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron, SGDClassifier, LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons
from plot_region import *
```

```python
X,y = make_moons(n_samples=300, noise=0.3, random_state=0)
```

```python
clr5 = LinearSVC()
scalar = StandardScaler()
X_sc = scalar.fit_transform(X)
clr5.fit(X_sc, y)
PlotResultRegion(clr5, X_sc, y, h = 0.02)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
perc = Perceptron(max_iter=1000, tol=0.001)
perc.fit(X_train,y_train)
yperc = perc.predict(X_test)
accuracy_score(yperc, y_test)
```

```python
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(X_train,y_train)
yperc = logreg.predict(X_test)
accuracy_score(yperc, y_test)
```

```python
linsvc = LinearSVC()
linsvc.fit(X_train,y_train)
yperc = linsvc.predict(X_test)
accuracy_score(yperc, y_test)
```

```python
svc = SVC(kernel='rbf', gamma=50)
svc.fit(X_train,y_train)
yperc = svc.predict(X_test)
accuracy_score(yperc, y_test)
```

```python
clr5 = SVC(kernel='rbf', gamma=50)
clr5.fit(X, y)
PlotResultRegion(clr5, X, y, h = 0.02)
```

```python
clr5 = SVC(kernel='rbf', gamma=0.3)
clr5.fit(X, y)
PlotResultRegion(clr5, X, y, h = 0.02)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
clfrs = [KNeighborsClassifier(n_neighbors=5),GaussianNB(),
         DecisionTreeClassifier(max_depth=2),Perceptron(max_iter=1000, tol=1e-3),
        LogisticRegression(multi_class='auto',solver='liblinear')]
for clfr in clfrs:
    clfr.fit(X, y)
    PlotResultRegion(clfr, X, y, h = 0.02)
    yperc = clfr.predict(X_test)
    print(clfr.__class__.__name__,accuracy_score(yperc, y_test))
```
