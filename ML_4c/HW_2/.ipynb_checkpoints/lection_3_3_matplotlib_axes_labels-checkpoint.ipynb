```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,101)
```

```python
plt.xlim(-2, 12) #устанавливаем пределы по осям
plt.ylim(-1.5, 1.5)
plt.plot(x, np.sin(x))
```

Ось ОХ в другую сторону

```python
plt.xlim(12, -2) #пределы в обратном порядке
plt.ylim(-1.5, 1.5)
plt.plot(x, np.sin(x)) #ось OX в другую сторноу
```

Заголовок, легенда, подписи осей

```python
plt.style.use('seaborn-whitegrid') #отображение сетки
plt.xlim(-2, 12) 
plt.ylim(-1.5, 1.5)
plt.title('Пример 1') #заголовок
plt.xlabel('x') #подписи к оси x
plt.ylabel('y=sin(x)') #подписи к оси y
plt.plot(x, np.sin(x), label='sin(x)') #отображение информации в легенду
plt.legend()
```

Размер графика

```python
plt.figure(figsize=(14,10)) #размер графика
plt.xlabel('x', fontsize=18)
plt.ylabel('y=sin(x)', fontsize=18)
plt.style.use('seaborn-whitegrid')
plt.plot(x, np.sin(x), label='sin(x)') 
```

Произвольный текст на графике, содержащий формулу

```python
import seaborn as sns
#plt.rc('text', usetex=True) если есть соответствующие пакеты теха. Запрашивает установить type1cm
sns.set(font_scale=1.6)#масштаб текста
sns.set_style("whitegrid")
plt.figure(figsize=(14,10))
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.xticks(np.arange(0, 10, step=1)) #тики по оси X
plt.plot(x, np.sin(x), label='sin(x)') 
plt.legend()
plt.text(1, -0.65, r'$\sum_{n=0}^\infty (-1)^n \frac{x^{2n+1}}{(2n+1)!}$', fontsize=20)
```
