```python
import numpy as np
import matplotlib.pyplot as plt
```

meshgrid - создание сетки

```python
x = np.array([1, 2, 3])
y = np.arange(4, 7, 1)
print(x)
print(y)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
print(xx)
print(yy)
print(xx)
```

```python
#будет рассмотрено на следующем слайде
print(xx.ravel()) #перевод в одномерный массив
print(yy.ravel()) #перевод в одномерный массив
np.c_[xx.ravel(), yy.ravel()] #составление пар
print(xx)
```

```python
plt.figure(figsize=(5,5))
plt.plot(xx, yy, marker='.', color='k', linestyle='none')
```

```python

zz = np.array([[0, 0, 0],[0, 1, 0],[1, 1, 1]])
```

```python
print(zz)
```

Функция contour

```python
plt.figure(figsize=(5,5))
cs = plt.contour(xx, yy, zz)
```

```python
x = np.arange(1, 3.05, 0.05)
y = np.arange(4, 6.05, 0.05)
len(x),len(y)
```

```python
xx, yy = np.meshgrid(x, y)
zz = np.eye(41,41)
print(zz)
```

```python
plt.figure(figsize=(5,5))
cs = plt.contour(xx, yy, zz)
plt.clabel(cs) #отображение подписей к линиям уровней
```

```python
plt.figure(figsize=(5,5))
plt.contourf(xx, yy, zz, alpha=0.4) #Области с заливкой
```

```python
plt.figure(figsize=(5,5))
cs = plt.contour(xx, yy, zz, levels=[0.2, 0.5, 1]) #отображение только заданных уровней
plt.clabel(cs)
```
