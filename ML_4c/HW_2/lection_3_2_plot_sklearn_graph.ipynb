```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Разбиваем тестовую выборку на 3 массива, в зависимости от класса. Берём только две характеристики: Длина чашелистика, Длина лепестка

```python
X0 = [[x_i[0], x_i[2]] for x_i,y_i in zip(X_train, y_train) if y_i == 0]
X1 = [[x_i[0], x_i[2]] for x_i,y_i in zip(X_train, y_train) if y_i == 1]
X2 = [[x_i[0], x_i[2]] for x_i,y_i in zip(X_train, y_train) if y_i == 2]
```

```python
n=5
X_testp = [[x_i[0],x_i[2]] for x_i in X_test[:n]]
```

Для рисования класса Щетинистый используется plot с опцией linestyle="", для остальных классов - scatter

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.6)
sns.set_style("whitegrid")
plt.figure(figsize=(10,8))
plt.plot([x[0] for x in X0], [x[1] for x in X0], linestyle="", color='red', marker='^', label='Щетинистый (0)')
plt.scatter([x[0] for x in X1], [x[1] for x in X1], color='green', marker='x', label='Разноцветный (1)')
plt.scatter([x[0] for x in X2], [x[1] for x in X2], color='blue', marker='x', label='Виргинский (2)')
plt.scatter([x[0] for x in X_testp], [x[1] for x in X_testp], color='black', marker='s', label='Новые экземпляры')
plt.xlabel('Длина чашелистика')
plt.ylabel('Длина лепестка')
#plt.legend(loc='lower right')
plt.show()
```
