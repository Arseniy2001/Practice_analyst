### Загрузка изображения из файла. Python Imaging Library (PIL)

```python
from PIL import Image
jpg = Image.open("picture.jpg")
```

```python
type(jpg)
```

```python
plt.imshow(jpg)
```

```python
import numpy as np
im2arr = np.array(jpg)
```

```python
im2arr.shape
```

```python
im2arr[0,0]
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.95870, 0.1140])
```

```python
jpgBW = rgb2gray(im2arr)
print(jpgBW.shape)
print(jpgBW.dtype)
```

```python
jpgBW1dim = jpgBW.flatten()
jpgBW1dim.shape
```

```python
plt.imshow(jpgBW, cmap=plt.get_cmap('gray'))
```

```python
arr2im = Image.fromarray(jpgBW)
```

```python
arr2im.convert('L').save('pictureBW.jpeg')
```
