```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,101) #равномерно разбиваем отрезок [0;10] 101й точкой (включая 0 и 10)
print(x[:5], len(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x+np.pi/4))
plt.plot(x, np.cos(x+np.pi/4))
```

```python
plt.plot(x, np.sin(x), linestyle='solid', color='green')
plt.plot(x, np.cos(x), linestyle='dashed', color='blue')
plt.plot(x, np.sin(x+np.pi/4), linestyle='dashdot', color='lightgreen')
plt.plot(x, np.cos(x+np.pi/4), linestyle='dotted', color='lightblue')
```


### короткие обозначения solid -> '-' dashed -> '--' dashdot -> '-.' dotted -> ':'

```python
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.cos(x), '--b')
plt.plot(x, np.sin(x+np.pi/4), '-.', 'lightgreen')
plt.plot(x, np.cos(x+np.pi/4), ':', 'lightblue')
```

### Сохраним значения функций в переменные

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x + np.pi/4)
y4 = np.cos(x + np.pi/4)
print(x[:5], len(x), y1[:5], len(y1))
```

```python
plt.plot(x, y1, marker='o', color='green')
plt.plot(x, y2, '.', color='red', linestyle='-') #с соединительной линией
plt.plot(x, y3, 's', color='blue')
plt.plot(x, y4, '^', color='orange')
```

### scatter plot (диаграмма рассеяния) позволяет задать каждой точке свой цвет и размер

```python
N = 45
x, y = np.random.rand(2, N)
c = np.random.randint(1, 5, size=N)# цвет
s = np.random.randint(40, 220, size=N)# размер

plt.scatter(x, y, c=c, s=s)
plt.colorbar();  # show color scale
c
```
