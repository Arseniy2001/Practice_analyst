```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.set_printoptions(precision=4, suppress=True, threshold=5)
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
X[0]
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
```

```python
pca.mean_
```

```python
X_m = X-X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_m)
```

```python
Vt
```

```python
np.matrix(Vt).H * np.matrix(Vt)
```

```python
sig = np.zeros([200,2])
sig[0,0]=s[0]
sig[1,1]=s[1]
```

```python
s
```

```python
mult = np.dot(U,sig)
mult = np.dot(mult,Vt)
```

```python
mult-X+pca.mean_
```

```python
V=np.matrix(Vt).H
```

```python
Vt
```

```python
Vm = V[:,0]
Vm
```

```python
X_new = np.dot(X_m, Vm)
```

```python
X_new
```

```python
import math
from matplotlib.pyplot import figure
figure(figsize=(6, 6), dpi=100)
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)

plt.scatter(X[:, 0], X[:, 1])
x0,y0 = pca.mean_

plt.scatter(x0,y0, color='r')

for exp_var, component in zip(pca.explained_variance_, pca.components_):
    dx, dy = math.sqrt(exp_var) * component
    plt.arrow(x0,y0,dx,dy,length_includes_head = True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.33, random_state=4)
```

```python
pca = PCA()
logistic = LogisticRegression(max_iter=10000, tol=0.1)
pipeline = Pipeline(steps=[('pca', pca), ('logistic', logistic)])
```

```python
param_grid = {
    'pca__n_components': [1,2,3,4],
    'logistic__C': np.logspace(-4, 4, 9),
}
```

```python
search = GridSearchCV(pipeline, param_grid, n_jobs=-1)
search.fit(X_train, y_train)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```
