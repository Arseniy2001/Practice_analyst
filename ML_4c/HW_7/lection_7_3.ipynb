```python
from sklearn import datasets
from sklearn.cluster import KMeans,DBSCAN
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++').fit(X)
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
X[kmeans.labels_ == 1][:,0].mean()
```

```python
X[kmeans.labels_ == 1][:3]
```

```python
x,y = np.hsplit(X, 4)[0:3:2]
```

```python
x0,x2 = np.hsplit(X[kmeans.labels_ == 1], 4)[0:3:2]
```

```python
sns.set(font_scale=1.6)
sns.set_style("whitegrid")
plt.figure(figsize=(10,8))
plt.scatter(x[kmeans.labels_ == 0],y[kmeans.labels_ == 0], color='red', marker='x', label='(0)')
plt.scatter(x[kmeans.labels_ == 1],y[kmeans.labels_ == 1], color='green', marker='x', label='(1)')
plt.scatter(x[kmeans.labels_ == 2],y[kmeans.labels_ == 2], color='blue', marker='x', label='(2)')
#plt.scatter(x[kmeans.labels_ == 3],y[kmeans.labels_ == 3], color='brown', marker='x', label='(3)')
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,2], color=['r','g','b','brown'], s=270)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,2], color=['r','g','b'], s=270)
plt.xlabel('Длина чашелистика')
plt.ylabel('Длина лепестка')
plt.legend(loc='lower right')
plt.show()
```

```python
inertia = [KMeans(n_clusters=i).fit(X).inertia_ for i in range(1,11)]
```

```python
plt.plot(range(1,11), inertia)
plt.xticks(range(1,11))
```

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=300, noise=0.08, random_state=0)
```

```python
x[:3] 
```

```python
kmeans3 = KMeans(n_clusters=2)
kmeans3.fit(x)
```

```python
plt.scatter(x[:,0][kmeans3.labels_ == 0],x[:,1][kmeans3.labels_ == 0], color='red', marker='x', label='(0)')
plt.scatter(x[:,0][kmeans3.labels_ == 1],x[:,1][kmeans3.labels_ == 1], color='green', marker='x', label='(1)')
#plt.scatter(x[kmeans.labels_ == 2],y[kmeans.labels_ == 2], color='blue', marker='x', label='(2)')
#plt.scatter(x[kmeans.labels_ == 3],y[kmeans.labels_ == 3], color='brown', marker='x', label='(3)')
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,2], color=['r','g','b','brown'], s=270)
plt.scatter(kmeans3.cluster_centers_[:,0],kmeans3.cluster_centers_[:,1], color=['r','g'], s=270)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.21).fit(x)
dbscan.labels_
```

```python
centroids = []
for lbl in np.unique(dbscan.labels_):
    points_of_cluster = x[dbscan.labels_==lbl,:]
    centroid_of_cluster = np.mean(points_of_cluster, axis=0) 
    centroids.append(centroid_of_cluster)
```

```python
centroids = np.array(centroids)
centroids[:,0]
```

```python
plt.scatter(x[:,0][dbscan.labels_ == 0],x[:,1][dbscan.labels_ == 0], color='red', marker='x', label='(0)')
plt.scatter(x[:,0][dbscan.labels_ == 1],x[:,1][dbscan.labels_ == 1], color='green', marker='x', label='(1)')
plt.scatter(centroids[:,0],centroids[:,1], color=['r','g'], s=270)
plt.show()
```
